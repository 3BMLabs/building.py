```python
import ifcopenshell
import ifcopenshell.geom
import ifcopenshell.util.element
import ifcopenshell.util.shape

model = ifcopenshell.open('C:/Users/Jonathan/Documents/GitHub/building.py/sandbox/IFC/models/Revit_Export_2257_4X3.ifc')
```

```python
projectdata = model.by_type("IfcProject")
print(projectdata)
```

```python
levels = model.by_type("IfcBuildingStorey")
print(levels)
```

```python
if levels:
    first_building_storey = levels[0]

    relations = model.get_inverse(first_building_storey)
    relation_objects = [rel for rel in relations if rel.is_a("IfcRelAggregates") or rel.is_a("IfcRelContainedInSpatialStructure")]
    print(relation_objects)

    for relation in relation_objects:
        related_elements = relation.RelatedElements if relation.is_a("IfcRelContainedInSpatialStructure") else relation.RelatedObjects
        for element in related_elements:
            print(element)
else:
    print("No building storeys found in the model.")
```
